```python
import osmnx as ox

place = "Kumaraguru College of Technology, Coimbatore, India"
G = ox.graph_from_place(
    place,
    network_type="walk",  # or "drive", "bike"
    simplify=True
)
```

```python
import rasterio

with rasterio.open("data/srtm.tif") as src:
    print("CRS:", src.crs)
    print("Bounds:", src.bounds)
    print("Shape:", src.shape)
    print("Count (bands):", src.count)
```

```python
G = ox.elevation.add_node_elevations_raster(
    G,
    filepath="data/srtm.tif"
)
```

```python
G = ox.elevation.add_node_elevations_google(
    G,
    api_key=None   # OSMnx uses free Open-Elevation fallback
)
```

```python
G = ox.elevation.add_edge_grades(G, add_absolute=True)
```
