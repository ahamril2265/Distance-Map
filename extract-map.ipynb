```python
import osmnx as ox

place = "Kumaraguru College of Technology, Coimbatore, India"
G = ox.graph_from_place(
    place,
    network_type="walk",  # or "drive", "bike"
    simplify=True
)
```

```python
G = ox.elevation.add_node_elevations_raster(
    G,
    filepath="data/srtm.tif"
)
```

```python
G = ox.elevation.add_edge_grades(G, add_absolute=True)
```
